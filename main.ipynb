# Data Cleaning

***Change Data Types***

```python
# discounted_price is in object type because of rupee sign
df['discounted_price'] = df['discounted_price'].str.replace("‚Çπ", "")
df['discounted_price'] = df['discounted_price'].str.replace(",", "")
```

```python
df['discounted_price']=df['discounted_price'].astype('float64')
```

```python
# actual_price also in object type because of rupee sign

df['actual_price'] = df['actual_price'].str.replace("‚Çπ", "")
df['actual_price'] = df['actual_price'].str.replace(",", "")
```

```python
df['actual_price']=df['actual_price'].astype('float64')
```

```python
df.info()
```

```python
df.describe()
```

```python
# discount_percentage

df['discount_percentage'] = df['discount_percentage'].str.replace("%", "").astype('float64')
```

```python
df['rating_count'] = df['rating_count'].str.replace(",", "").astype('float64')
```

```python
# rating

df.rating.value_counts()
```

```python
# '|' beacuse of this symbol rating cant be float

df.query('rating=="|"')
# df[df['rating']== "|"]
```

```python
df['rating'] = df['rating'].str.replace("|", '3.5')
df['rating'] = df['rating'].astype('float64')
```

```python
df.info()
```

```python
# Descriptive Stats

df.describe()
```

```python
df.rating_count.isnull().sum()
```

```python
df[df['rating_count'].isnull()]
```

```python
print("mean: ",df['rating_count'].mean())
print("median: ", df['rating_count'].median())
```

```python
# replacing rating_count Null values with median 
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
```

```python
# to check if there are any duplicates

df.duplicated().any()
```

```python
plt.figure(figsize=(12,6))
plt.title('Main Category Distibution') 
ax = sns.countplot(x=df['main_category'], hue = df['main_category'], palette='Set2')
ax.bar_label(ax.containers[0])
plt.xlabel('Main_Category')
plt.ylabel('No of Products')
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
```

```python
bins =[0,1,2,3,4,5]

df['cluster_rating'] = pd.cut(df['rating'], bins=bins, labels= ['0-1','1-2','2-3','3-4','4-5'])
```

```python
df.cluster_rating.value_counts().sort_index()
```

***
Observation:

1. Majority ratings fall between 3-4 and 4-5 range.
2. There is a slight increase in ratings 2-3 compared to 1-2 and 0-1 ranges.
3. Means overall customers are satisfied with the products.
***

```python
topCategories_byRating = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
```
